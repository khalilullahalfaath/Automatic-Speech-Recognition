```python
input_path = '/content/drive/MyDrive/Kuliah/S-2 Universitas Gadjah Mada/Semester 1/RFPP/Tugas 2/Rekaman Tugas 2'
output_root = '/content/drive/MyDrive/Kuliah/S-2 Universitas Gadjah Mada/Semester 1/RFPP/Tugas 2/Rekaman Tugas 2/Sudah dipotong'

os.makedirs(output_root, exist_ok=True)

audio_files = sorted(glob(os.path.join(input_path, '*.m4a')))
print(f"Prefetched {len(audio_files)} files from Drive cache.")

def energy_based_split(y, sr, frame_length=2048, hop_length=512, energy_threshold=0.08, min_silence=0.25):
    rms = librosa.feature.rms(y=y, frame_length=frame_length, hop_length=hop_length)[0]
    frames = np.nonzero(rms > energy_threshold * np.max(rms))[0]
    if len(frames) == 0:
        return np.array([[0, len(y)]])
    times = librosa.frames_to_samples(frames, hop_length=hop_length)
    segs = []
    start = times[0]
    for i in range(1, len(times)):
        if times[i] - times[i-1] > min_silence * sr:
            segs.append([start, times[i-1]])
            start = times[i]
    segs.append([start, times[-1]])
    return np.array(segs)

def auto_cut_audio(input_file, output_dir, top_db=40, min_duration=0.25):
    # Load audio with robust backend
    y, sr = librosa.load(input_file)

    # Clean low-frequency noise
    b, a = signal.butter(4, 100/(sr/2), btype='highpass')
    y = signal.filtfilt(b, a, y)

    # Try default splitting
    intervals = librosa.effects.split(y, top_db=top_db)

    # Adaptive fallback
    if len(intervals) <= 1:
        for new_db in [45, 50, 55, 60]:
            intervals = librosa.effects.split(y, top_db=new_db)
            if len(intervals) > 1:
                break
        if len(intervals) <= 1:
            intervals = energy_based_split(y, sr, energy_threshold=0.08)

    # Save each segment to folder
    os.makedirs(output_dir, exist_ok=True)
    base = Path(input_file).stem
    saved = 0
    for i, (start, end) in enumerate(intervals, start=1):
        duration = (end - start) / sr
        if duration < min_duration:
            continue
        segment = y[start:end]
        out_path = Path(output_dir) / f"{base}_seg_{i}.wav"
        sf.write(out_path, segment, sr)
        saved += 1

    print(f"{base}: {saved} segmen disimpan.")
    return saved

for file in audio_files:
    name = Path(file).stem

    # Create one folder per file to trigger Drive metadata refresh
    target_dir = Path(output_root) / name
    os.makedirs(target_dir, exist_ok=True)

    count = auto_cut_audio(file, target_dir, top_db=40)

    # Force FUSE to flush metadata immediately
    os.sync()
    !sync
    print(f"✅ Synced {name} → Drive folder")

print("=== Semua file selesai dan sudah tersinkron dengan Drive ===")
```
